## Performance analysis

Contrary to what might have been expected, the standard MCTS has a better performance than all the modified versions that we implemented. This result can be explain by the following factores:

- In case of the standard version, the exploration-exploitation balance of the UCT formula is already well calibrated for general game-playing scenarios. When we add constraints such as avoiding repeated states (V2) or replacing random rollouts with heuristic ones (V3), we are introducing assumptions about what constitutes a "good" play that may not always hold in PopOut specifically.

- In the case of V2, avoiding repeated states reduces the number of moves available at each node, which can make the tree less diverse and cause 
the algorithm to miss certain winning lines that involve revisiting states.

- In the case of V3, the heuristic rollouts are more expensive to compute, meaning that with the same number of iterations the algorithm explores fewer paths than V1. Since the heuristic only looks one move ahead, it may also introduce a bias that is not always beneficial in a game as dynamic as PopOut, where the pop move can completely change the board in unexpected ways.

- In the case of V4 (Memory + Heuristic MCTS), which combines the strengths of V2 and V3, the performance is also lower than the standard 
version. This is because it inherits the limitations of both — the reduced diversity of the search tree from V2 and the higher computational 
cost per iteration from V3. While in theory combining these two approaches should produce better results, in practice the reduced number of paths explored due to both constraints working simultaneously makes the algorithm less effective than the standard version, at least with the number of iterations used in our tests.
