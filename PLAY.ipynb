#### **8. Playing the game** <br>
<div style="text-align: justify;"><p style="text-indent: 2em;">This notebook defines the menus for the game execution in the terminal. The first two functions, "select_simulations" and "select_algorithm", are responsible for asking the player's input about the computers' search algorithm to be used and, if applicable, the number of simulations that algorithm should perform.</div>

```python
import sys
import time
%run PopOut.ipynb
%run AuxiliarFunctions.ipynb 

def select_simulations():
    num_simulations = None
    while not num_simulations:
        clear_terminal()
        print("Introduce a number of simulations between 10 and 1000:")
        try: 
            num_simulations = int(input())
            if (num_simulations < 10 or num_simulations > 1000): 
                num_simulations = None
                clear_terminal()
                print("Number outside of selection range. Returning to SIMULATIONS SELECTION...")
                time.sleep(2)
        except:
            clear_terminal()
            print("Not a readable number. Returning to SIMULATIONS SELECTION...")
            time.sleep(2)
    return num_simulations

def select_algorithm(num = 1):
    algorithm = None
    num_simulations = None
    while not algorithm:
        clear_terminal()
        print(f"Select algorithm for COMPUTER PLAYER {num}:\n\n1. Standard Monte Carlo\n2. Move Memory Monte Carlo\n3. Heuristic Monte Carlo\n4. Move Memory + Heuristic Monte Carlo\n5. Decision Trees with ID3")
        match(input().strip()):
            case "1": algorithm = "standard MCTS"
            case "2": algorithm = "memory MCTS"
            case "3": algorithm = "heuristic MCTS"
            case "4": algorithm = "memory+heuristic MCTS"
            case "5": return "decision trees", None
            case _:
                clear_terminal()
                print("Invalid command! Returning to ALGORITHM SELECTION...")
                time.sleep(2)
    num_simulations = select_simulations()
    return algorithm, num_simulations   
```

<div style="text-align: justify;"><p style="text-indent: 2em;">Then, the following while cycle allows for navigation in menus. We have a starting menu, which allows to read rules and commands for Pop Out, play the game or quit the execution. Then, if the user proceeds with playing the game, a second menu asks for input about game mode (player vs player, player vs computer or computer vs computer). If there is a computer player in that game mode, then functions to select algorithm and simulations are called.</div>

```python
while True:
    clear_terminal()
    print("Welcome to POP OUT!\n\n1. CHECK COMMANDS\n2. CHECK RULES\n3. PLAY\n4. EXIT")
    match(input().strip()):    
        case "1": check_commands()                                  # Shows the available commands. 
        case "2": check_rules()                                     # Shows the rules of the game.
        case "3":
            while True:
                clear_terminal()
                print("Select game mode:\n\n1. PLAYER VS PLAYER\n2. PLAYER VS COMPUTER\n3. COMPUTER VS COMPUTER\n4. RETURN TO MENU")
                match(input().strip()):
                    case "1":                                       # Starts game between two human players.  
                        game = Pop_Out()
                        while not game.terminal: PVP_play(game)
                        win(game)
                    case "2":                                       # Starts game between a human and a computer.
                        algorithm, num_simulations = select_algorithm()
                        game = CVP_Pop_Out()
                        while not game.terminal: PVC_play(game, algorithm, num_simulations)
                        win(game)
                    case "3":                                       # Starts game between two computers.
                        algorithm1, num_simulations1 = select_algorithm(1)
                        algorithm2, num_simulations2 = select_algorithm(2)
                        game = Pop_Out()
                        a = 1
                        while not game.terminal: 
                            if a == 0: CVC_play(game, algorithm1, num_simulations1)
                            if a == 1: CVC_play(game, algorithm2, num_simulations2)
                            a = (a+1)%2
                        win(game)
                    case "4": break                                 # Returns to main menu.
                    case _:                                         # Handles other unrecognized commands.
                        clear_terminal()
                        print("Invalid command! Returning to GAME SELECTION...")
                        time.sleep(2)
        case "4":                                                   # Terminates execution.
            sys.exit("Goodbye!")
        case _:                                                     # Handles other unrecognized commands.
            clear_terminal()
            print("Invalid command! Returning to MENU...")
            time.sleep(2)

# use:           ipython -c "%run PLAY.ipynb"            in the terminal
```
